````python
import seaborn as sns
ax = sns.pointplot(x=previsao_de_renda['posse_de_imovel'], y=previsao_de_renda['renda'], errorbar=('ci', 95), dodge=True)
````

````python
# outras forma de fazer a analise de significancia de 95% de confiança.

import numpy as np


for col in ['posse_de_imovel', 'posse_de_veiculo']:
    #filtrar renda entre true e false
    renda_true = previsao_de_renda[previsao_de_renda[col] == 1]['renda']
    renda_false = previsao_de_renda[previsao_de_renda[col] == 0]['renda']
    
    # Tamanho da amostra
    n_true = len(renda_true)
    n_false = len(renda_false)
    
    # Média da amostra
    media_true = np.mean(renda_true)
    media_false = np.mean(renda_false)
    
    # Desvio padrão da amostra
    desvio_padrao_true = np.std(renda_true, ddof=1)
    desvio_padrao_false = np.std(renda_false, ddof=1)
    
    # Erro padrão da média
    erro_padrao_true = desvio_padrao_true / np.sqrt(n_true)
    erro_padrao_false = desvio_padrao_false / np.sqrt(n_false)
    
    # Valor crítico do z para um intervalo de confiança de 95%
    z_critico = 1.96
    
    # Limites do intervalo de confiança
    limite_inferior_true = media_true - z_critico * erro_padrao_true
    limite_superior_true = media_true + z_critico * erro_padrao_true
    
    limite_inferior_false = media_false - z_critico * erro_padrao_false
    limite_superior_false = media_false + z_critico * erro_padrao_false
    
    print(f"95% confidence interval for the mean {col} true: ({limite_inferior_true}, {limite_superior_true})")
    print(f"95% confidence interval for the mean {col} false: ({limite_inferior_false}, {limite_superior_false})")
````

Teste t de Student para posse de imóvel e posse de veículo em relação à renda   
obs: o teste t de Student é um teste estatístico que é utilizado para verificar se há diferença significativa entre as médias de duas amostras independentes.
via de regra, se o p-valor for menor que 0.05, rejeitamos a hipótese nula, ou seja, as médias são diferentes.

````python
from scipy import stats

# Realizando um teste t de Student para posse de imóvel
t_statistic, p_value = stats.ttest_ind(previsao_de_renda[previsao_de_renda['posse_de_imovel'] == 1]['renda'], previsao_de_renda[previsao_de_renda['posse_de_imovel'] == 0]['renda'])
print(f"t-statistic: {t_statistic}")
print(f"p-value: {p_value}")
````

````python
# Realizando um teste t de Student para posse de veículo
t_statistic, p_value = stats.ttest_ind(previsao_de_renda[previsao_de_renda['posse_de_veiculo'] == 1]['renda'], previsao_de_renda[previsao_de_renda['posse_de_veiculo'] == 0]['renda'])
print(f"t-statistic: {t_statistic}")
print(f"p-value: {p_value:.15f}")
````

Conclusão: As Duas variáveis são extremamente significativas para prever a renda, pois o p-valor é bem menor que 0.05, e o intervalo de confiança de 95% estão bem distantes ou seja, as médias são diferentes.
