1- Calcule de forma numérica a integral da função $x^3$ avaliada entre os pontos $1$ e $2$. Identifique o número de retângulos que você usou.  


```python
import matplotlib as mpl
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import warnings
import numpy as np
warnings.filterwarnings("ignore")

# Função quadrática
f = lambda x: x**2

def calcula_area(a, b, func, frame_num):
    num_bins = frame_num + 1  # Quantidade de retângulos
    delta = (b - a) / num_bins  # Largura do retângulo
    x_cols = np.linspace(a, b - delta, num_bins) + delta / 2  # Divisão do eixo X
    y_cols = f(x_cols)  # Alturas dos retângulos
    return sum(delta * y_cols)

def animate(n, a=1, b=2, func=f, verdadeiro_valor=(((1/3)*2**3)-(1/3)*1**3)):
    # Limpa o conteúdo do gráfico para o próximo frame
    ax.cla()
    
    # Define os limites do gráfico
    ax.set_xlim(0, b + 1)  # Curva mostrada de 0 a 2
    ax.set_ylim(0, f(b+1))

    # Plota a função de x=0 até x=2
    x = np.linspace(0, b+1, 100)  # Aqui, o intervalo de 0 até b para a curva completa
    y = f(x)
    ax.plot(x, y, '-')

    # Define as quebras e alturas das colunas apenas de x=1 até x=2
    num_bins = n + 1
    delta = (b - a) / num_bins
    x_cols = np.linspace(a, b - delta, num_bins) + delta / 2
    y_cols = f(x_cols)

    # Plota as barras entre x=1 e x=2
    ax.bar(x_cols, y_cols, alpha=0.5, width=delta, edgecolor='blue')   

    # Calcula a área dos retângulos e o erro
    soma_riemann = calcula_area(a, b, func, n)
    erro = verdadeiro_valor - soma_riemann
    ax.set_title(f'Soma de Riemann = {soma_riemann:.4f} erro = {erro:.5f}')


# resolver o problema de não mostrar o gráfico
mpl.use('Agg')

fig, ax = plt.subplots(figsize=(10, 6))

# Cria a animação
ani = FuncAnimation(fig, animate, frames=40, interval=120, blit=False) # 40 frames, portanto 40 retângulos!!

# Exibe a animação no notebook
display(HTML(ani.to_jshtml()))    
```
