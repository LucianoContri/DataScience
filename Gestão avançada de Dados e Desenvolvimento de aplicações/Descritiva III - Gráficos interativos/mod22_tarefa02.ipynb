#### 1. Carregue os dados da tarefa anterior

````python
import pandas as pd

# import the multi index data from the csv file
df = pd.read_csv('stock_data_tech.csv', header=[0, 1], index_col=0, parse_dates=True)
df.tail()
````

#### 2. Média móvel

1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu
2. Utilizando a fução ```rolling()```, calcule o desvio padrão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima
2. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
4. Monte um gráfico de linhas para os desvios padrão em janela móvel
5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

````python
import plotly.express as px
import plotly.graph_objects as go


# 1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu
# 2. Utilizando a fução ```rolling()```, calcule o desvio padrão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima
# 2. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
# 4. Monte um gráfico de linhas para os desvios padrão em janela móvel
# 5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

tickers = ['NVDA', 'AMD', 'INTC', 'QCOM']
# 1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu

for ticks in df.columns.levels[0]:
    df[ticks, 'Close_90'] = df[ticks, 'Adj Close'].rolling(window=90).mean() # média móvel  em 90 dias dos valores de fechamento ajustados
    df[ticks, 'Close_90_std'] = df[ticks, 'Adj Close'].rolling(window=90).std() # desvio padrão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima

# Step 2: Select all 'Close_90' columns
close_90_df = df.xs('Close_90', axis=1, level=1)

# Step 3: Plot the data
fig = go.Figure()

for ticker in close_90_df.columns:
    fig.add_trace(go.Scatter(
        x=close_90_df.index, 
        y=close_90_df[ticker], 
        mode='lines', 
        name=ticker
    ))

fig.update_layout(
    title='90-Day Moving Averages of Adjusted Closing Prices',
    xaxis_title='Date',
    yaxis_title='Price',)

fig.show()

#---------------------------------------------------------------------------------

# Select all 'Close_90_std' columns
std_90_df = df.xs('Close_90_std', axis=1, level=1)
# Plot the data
fig_std = go.Figure()

for ticker in std_90_df.columns:
    fig_std.add_trace(go.Scatter(
        x=std_90_df.index, 
        y=std_90_df[ticker], 
        mode='lines', 
        name=ticker
    ))

fig_std.update_layout(
    title='90-Day Rolling Standard Deviations of Adjusted Closing Prices',
    xaxis_title='Date',
    yaxis_title='Standard Deviation',)

fig_std.show()
````

#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
ticker = 'NVDA'

# Ensure the index is datetime
df.index = pd.to_datetime(df.index)

# Get the last date in the DataFrame
last_date = df.index.max()

# Calculate the date 59 days before the last date (for a total of 60 days)
start_date = last_date - pd.Timedelta(days=59)

# Select data for the ticker and date range
df_ticker = df.loc[start_date:last_date, ticker]

# Step 2: Prepare the data for the candlestick chart
df_candle = df_ticker[['Open', 'High', 'Low', 'Close']].dropna()

# Step 3: Create the candlestick chart
fig = go.Figure(data=[go.Candlestick(
    x=df_candle.index,
    open=df_candle['Open'],
    high=df_candle['High'],
    low=df_candle['Low'],
    close=df_candle['Close'],
    increasing_line_color='green',
    decreasing_line_color='red',
    name=ticker
)])

# Update the layout
fig.update_layout(
    title=f'Candlestick Chart for {ticker} - Last 60 Days',
    xaxis_title='Date',
    yaxis_title='Price',
    xaxis_rangeslider_visible=False
)

# Show the plot
fig.show()
````
